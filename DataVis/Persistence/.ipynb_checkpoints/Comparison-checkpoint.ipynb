```python
# Initial imports
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import datetime
from matplotlib import pyplot as plt
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('Dark2')
import statsmodels.api as sm
```

```python
# Import miranda shocks
shockdf = pd.read_excel('mirandause.xlsx')
shockdf.set_index('Announcement Date', inplace=True)
shockdf.index = shockdf.index.map(lambda dt: dt.replace(hour=0))
```

```python
# Import LIBOR 3m 
# Get data series from FRED
start = datetime.datetime(1997, 1, 1)
end = datetime.datetime(2018, 8, 20)
freddata = web.DataReader(['GBP3MTD156N'], 'fred', start, end)
freddata.rename(columns={"GBP3MTD156N": "Libor"}, inplace=True)
```

```python
# Merge dataframes into one
df = pd.concat([freddata, shockdf], axis=1)
```

```python
X = df.loc[:, 'SS1'].copy()
X = sm.add_constant(X, prepend=True)
y = df.loc[:, 'Libor'].copy().shift(-1)
X['lag'] = df.loc[:, 'Libor'].copy().shift(1)
```

```python
# Regressions of
    # MarketMoment_{t+s}  = const + beta_1 * MarketMoment_{t-1} + beta_2 * shock_{t}

leadno = 31

coeff_array = np.zeros(leadno)
t_array = np.zeros(leadno)
stderr_array = np.zeros(leadno)

for i in range(leadno):
    X = df.loc[:, 'SS1*'].copy()
    X = sm.add_constant(X, prepend=True)
    y = df.loc[:, 'Libor'].copy().shift(-i)
    X['lag'] = df.loc[:, 'Libor'].copy().shift(1)
    model = sm.OLS(y, X, missing='drop')
    res = model.fit()
    coeff_array[i] = res.params[1]
    t_array[i] = res.tvalues[1]
    stderr_array[i] = res.bse[1]
```

```python
# Plot IRFs

plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.plot(np.arange(leadno-1), coeff_array[1:], label='$\\beta_2$')
plt.plot(np.arange(leadno-1), coeff_array[1:]+1.64*stderr_array[1:], 'k:', label='90% CI')
plt.plot(np.arange(leadno-1), coeff_array[1:]-1.64*stderr_array[1:], 'k:')
plt.title('Impulse Response from a shock to the first moment a la Miranda-Agrippino (2016)')
plt.xlabel('Trading days (s)')
plt.legend()

plt.savefig('C:/Users/Tim/Documents/Nuffield/MphilThesis/DataVis/Persistence/Figs/Comparison.pdf')
plt.show()
```
