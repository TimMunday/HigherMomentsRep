```python
# Imports
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('tab20')
import os
directory = 'C:/Users/tmund/Documents/Submissions/HigherMomentsRep/'
os.chdir(directory)
```

```python
mypath = 'DataVis/CaseStudies/Figs/'
```

```python
# Import data

# financial data (purged)
m12df = pd.read_pickle('Output/2 - PurgedFinancial/m12df_chg_purged.pkl')
```

```python

m12df['StDevresid'].nlargest(5)
```

```python
m12df['StDevresid'].nsmallest(5)
```

```python
# Plot standard deviation
    # Highlight biggest move, restricted to where we have both Q and A's and an IR
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(m12df.StDevresid, label='Change in Residual Standard Deviation')
plt.plot(m12df['StDevresid'].nlargest(2).index[0],
         m12df['StDevresid'].nlargest(2)[0],
         color=palette(2),
         marker='.',
         markersize = 15.,
        label='November 2001')
plt.plot(m12df['StDevresid'].nlargest(2).index[1],
         m12df['StDevresid'].nlargest(2)[1],
         color=palette(3),
         marker='.',
         markersize = 15.,
        label='November 2010')
plt.plot(m12df['StDevresid'].nsmallest(2).index[0],
        m12df['StDevresid'].nsmallest(2)[0],
         color=palette(4),
         marker='.',
         markersize = 15.,
        label='August 2016')
plt.plot(m12df['StDevresid'].nsmallest(2).index[1],
        m12df['StDevresid'].nsmallest(2)[1],
         color=palette(5),
         marker='.',
         markersize = 15.,
        label='August 1999')
plt.legend()
plt.xlabel('Year')

# Plot skew
plt.subplot(1, 2, 2)
plt.plot(m12df.Skewresid, label='Change in Residual Skewness')
plt.plot(m12df['Skewresid'].nlargest(2).index[0],
         m12df['Skewresid'].nlargest(2)[0],
         color=palette(2),
         marker='.',
         markersize=15.0,
         label='May 2014')
plt.plot(m12df['Skewresid'].nlargest(2).index[1],
         m12df['Skewresid'].nlargest(2)[1],
         color=palette(3),
         marker='.',
         markersize=15.0,
         label='November 2001')
plt.plot(m12df['Skewresid'].nsmallest(2).index[0],
         m12df['Skewresid'].nsmallest(2)[0],
         color=palette(4),
         marker='.',
         markersize=15.0,
         label='February 2014')
plt.plot(m12df['Skewresid'].nsmallest(2).index[1],
         m12df['Skewresid'].nsmallest(2)[1],
         color=palette(5),
         marker='.',
         markersize=15.0,
         label='November 2010')
plt.xlabel('Year')
plt.legend()


plt.savefig(mypath +str('CaseStudies.pdf'))
plt.show()
```
