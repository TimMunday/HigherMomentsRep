```python
# Initial imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('Dark2')
from collections import namedtuple
import matplotlib.ticker as ticker
import os
```

```python
mypath = 'C:/Users/Tim/Documents/Nuffield/MphilThesis/DataVis/PDFs/Figs/'
```

```python
# Import the data
m12df = pd.read_pickle('C:/Users/Tim/Documents/Nuffield/MphilThesis/Data/Pickled/m12df.pkl')
```

```python
# Plot 12 month ahead lines for visulation purposes

plt.figure(figsize=(20, 6))

plt.subplot(1, 3, 1)
plt.plot(m12df.index, m12df.Mean)
plt.title('Mean')

plt.subplot(1, 3, 2)
plt.plot(m12df.index, m12df.StDev)
plt.title('Standard Deviation')

plt.subplot(1, 3, 3)
plt.plot(m12df.index, m12df.Skew)
plt.title('Skew')

plt.savefig(mypath+str('moments.pdf'))

plt.show()
```
