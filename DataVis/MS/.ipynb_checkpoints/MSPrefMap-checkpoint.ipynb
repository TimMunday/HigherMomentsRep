```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('seaborn-darkgrid')
palette = plt.get_cmap('tab20c')
import ar1_approx as ar1
import scipy.stats as stats
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
import numba
from numba import jit
```

```python
mypath = 'C:/Users/Tim/Documents/Nuffield/MphilThesis/DataVis/MS/Figs/'
```

```python
# Create state space grid and put in parameters (x and y)
theta = 0.
size_x = 25
size_y = 25
sigma_x = 0.1
sigma_y = 0.2

alpha = 1/(sigma_y**2)
beta = 1/(sigma_x**2)

x_grid, x_pi_dontuse = ar1.addacooper(size_x, mu=theta, rho=0, sigma=sigma_x)
y_grid, y_pi_dontuse = ar1.addacooper(size_y, mu=theta, rho=0, sigma=sigma_y)
```

```python
# Create expectation given current grid points

def Exp_abar(x, y, a_grid_y):
    ''' Function to return expected a_bar
    Inputs:
        x: observed x (x_i); float
        y: observed y; float
        a_grid_y: The current vector in the a_grid corresponding to the observed y; numpy array
    Outputs:
        Expectation of a_bar; float'''
    xj_mean = (alpha*y+beta*x)/(alpha + beta)
    probs=np.zeros(len(x_grid))
    midpoints = np.zeros(len(x_grid)+1)
    midpoints[0] = -1000
    midpoints[len(x_grid)] = 1000
    for i in range(len(x_grid)-1):
        midpoints[i+1] = (x_grid[i] + x_grid[i+1])/2
    for i in range(len(x_grid)):
        probs[i] = (stats.norm.cdf(midpoints[i+1], loc=xj_mean, scale=sigma_x)
                     - stats.norm.cdf(midpoints[i], loc=xj_mean, scale=sigma_x))
    expectation = np.sum(probs*a_grid_y) # matrix multiplied by correct vector of a's in a grid
    return expectation

# r function
def r_func(y):
    '''Gives r as a function of public signal y
    To fit stylised facts, need r high in booms, low in slumps
    Inputs:
        y: observed y; float
    Ouputs:
        r: float'''
    r = stats.logistic.cdf(-y, loc=theta, scale=0.1)
    return r
    
# Evaluate a_i
def a_i(x, y, a_grid_y):
    '''Evaluate a_i according to the first order condition'''
    a_i = (1-r_func(y))*(alpha*y + beta*x)/(alpha+beta) + r_func(y)*Exp_abar(x, y, a_grid_y)
    return a_i

# Morris and Shin closed form
def a_star_linear(x, y):
    '''Given some vector of x's will show the optimal
    choices according to linear result in MS. With exogenous r
    Inputs:
        y: float
        x: can be numpy array of floats'''
    a_star = (alpha*y + beta*(1-0.5)*x)/(alpha + beta*(1-0.5))
    return a_star

# Morris and Shin closed for with r_func
def a_star_linear_rfunc(x, y):
    '''Given some vector of x's will show the optimal
    choices according to linear result in MS. With endogenous r
    Inputs:
        y: float
        x: can be numpy array of floats'''
    a_star = (alpha*y + beta*(1-r_func(y))*x)/(alpha + beta*(1-r_func(y)))
    return a_star
```

```python
plt.plot(y_grid, r_func(y_grid))
plt.show()
```

```python
# MS 3d
a_3d_ms = np.zeros((len(x_grid), len(y_grid))) #initialise matrix
X, Y = np.meshgrid(x_grid, y_grid)
Z = (alpha*Y + beta*(1-0.5)*X)/(alpha + beta*(1-0.5))

fig = plt.figure(figsize=(12, 5))
ax = fig.add_subplot(121, projection='3d')
ax.set_facecolor('white') 
ax.plot_surface(X, Y, Z,
                rstride=1,
                cstride=1,
                linewidth=0,
                cmap=cm.Blues)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('a')
ax.set_title('r = 0.5')

# MS rfunc 3d

Z = a_star_linear_rfunc(X, Y)
ax = fig.add_subplot(122, projection='3d')
ax.set_facecolor('white') 
ax.plot_surface(X, Y, Z,
                rstride=1,
                cstride=1,
                linewidth=0,
                cmap=cm.Blues)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('A')
ax.set_title('r = f(y); (logistic)')
plt.savefig(mypath + str('preferences.pdf'))
plt.show()
```
