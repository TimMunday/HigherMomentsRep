```python
# Import data
df = pd.read_pickle('Data/Surveys/Reuters/Reutersdf.pkl')
df = df.loc[df.loc[:, '1YearAbsErr']<90, :] # get rid of ones that are really far off one year ahead estimate
```

```python
# Make Beta  plot

# Create series, such that infinite are replaced by the max

beta_stdev = 1/df.StDev
beta_stdev.iloc[np.where(beta_stdev==np.inf)]=beta_stdev.iloc[np.where(beta_stdev!=np.inf)].max()

beta_range = 1/df.Range
beta_range.iloc[np.where(beta_range==np.inf)]=beta_range.iloc[np.where(beta_range!=np.inf)].max()

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(beta_stdev, label='$\\beta$: Inverse survey variance')
plt.xlabel('Year')
plt.ylabel('$\\beta$')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(beta_range, label = '$\\beta$: Inverse survey range')
plt.xlabel('Year')
plt.ylabel('$\\beta$')
plt.legend()

plt.savefig('DataVis/MS/Figs/beta.pdf')
plt.show()
```
