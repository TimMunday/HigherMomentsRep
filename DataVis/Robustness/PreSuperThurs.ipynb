```python
# Import data

nonzerocoeffs_pre = np.load('Output/5 - Robustness/Purged/PreSupThu/nonzerocoeffs.npy')

mean_ir_pre = np.load('Output/5 - Robustness/Purged/PreSupThu/mean_ir.npy')
mean_qa_pre = np.load('Output/5 - Robustness/Purged/PreSupThu/mean_qa.npy')
mean_stat_pre = np.load('Output/5 - Robustness/Purged/PreSupThu/mean_stat.npy')
stdev_ir_pre = np.load('Output/5 - Robustness/Purged/PreSupThu/stdev_ir.npy')
stdev_qa_pre = np.load('Output/5 - Robustness/Purged/PreSupThu/stdev_qa.npy')
stdev_stat_pre = np.load('Output/5 - Robustness/Purged/PreSupThu/stdev_stat.npy')
skew_ir_pre = np.load('Output/5 - Robustness/Purged/PreSupThu/skew_ir.npy')
skew_qa_pre = np.load('Output/5 - Robustness/Purged/PreSupThu/skew_qa.npy')
skew_stat_pre = np.load('Output/5 - Robustness/Purged/PreSupThu/skew_stat.npy')

nonzerocoeffs_entire_pre = np.load('Output/5 - Robustness/Purged/PreSupThu/Entire/nonzerocoeffs.npy')

mean_entire_pre = np.load('Output/5 - Robustness/Purged/PreSupThu/Entire/mean_entire.npy')
stdev_entire_pre = np.load('Output/5 - Robustness/Purged/PreSupThu/Entire/stdev_entire.npy')
skew_entire_pre = np.load('Output/5 - Robustness/Purged/PreSupThu/Entire/skew_entire.npy')

```

```python
# Plot for second moment pre supthu

plt.figure(figsize=(20, 6))

# IR
plt.subplot(1, 3, 1)
plt.hist(stdev_ir_pre, bins=20, label='Distribution of topics chosen under permutation')
plt.axvline(nonzerocoeffs_pre[3], color='k', linestyle='dashed', label='No. of topics chosen using correct sample')
plt.xlabel('Number of topics selected by Elastic Net')
plt.ylabel('Frequency')
plt.title('IR')
plt.legend(loc='best')

# QA
plt.subplot(1, 3, 2)
plt.hist(stdev_qa_pre, bins=20, label='Distribution of topics chosen under permutation')
plt.axvline(nonzerocoeffs_pre[4], color='k', linestyle='dashed', label = 'No. of topics chosen using correct sample')
plt.xlabel('Number of topics selected by Elastic Net')
plt.ylabel('Frequency')
plt.title('Q and A')
plt.legend(loc='best')

# Stat
plt.subplot(1, 3, 3)
plt.hist(stdev_stat_pre, bins=20, label='Distribution of topics chosen under permutation')
plt.axvline(nonzerocoeffs_pre[5], color='k', linestyle='dashed', label='No. of topics chosen using correct sample')
plt.xlabel('Number of topics selected by Elastic Net')
plt.ylabel('Frequency')
plt.legend(loc='best')
plt.title('Statement')

plt.savefig(mypath+str('ENPermStDevPreSupThu.pdf'))

plt.show()
```
